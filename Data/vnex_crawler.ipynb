# 1. Tải và cài đặt một số thư viện cho crawler

```python
import numpy as np
import pandas as pd
import csv
import newspaper
from newspaper import Article as Atc
from newspaper import Source as Scr
```

# 2. Tải tất cả đường link và lọc ra những đường link chỉ chứa html:

```python
vnex_paper = newspaper.build('https://vnexpress.net/', language = 'vi', follow_meta_refresh = True, keep_article_html=True)
```

```python
size = vnex_paper.size()
print(f"Co tat ca {size} den tu https://vnexpress.net/")
```

```python
valid_url = np.array([],np.string_)
for i in range(size):
  vnex_url = vnex_paper.article_urls()[i]
  # fillter out only-html link
  temp = ''
  for j in range(len(vnex_url)):
    temp = vnex_url[len(vnex_url) - j - 1] + temp
    if temp == 'htm' or temp == 'html':
      valid_url = np.append(valid_url,vnex_url)
print(f"Xuat hien {np.shape(valid_url)[0]} bai bao trong vnexpress")
```

```python
star = np.array(['Date', 'Content', 'Source'],np.string_)
for i in range(np.shape(valid_url)[0]):
  vnex_url = Atc(valid_url[i],language = 'vi')
  vnex_url.download()
  vnex_url.parse()
  row = [vnex_url.meta_data['lastmod'],vnex_url.text,vnex_url.url]
  star = np.append(star, row, axis = 0)
```

# 3. Lưu lại dữ liệu với định dạng Date,Content,Source:

```python
star = np.reshape(star,(star.size//3,3))
filename = 'vnex_vi.csv'
```

```python
with open(filename, 'a') as csvFile:
    csvWriter = csv.writer(csvFile, delimiter='|', quoting= csv.QUOTE_ALL)
    csvWriter.writerows(star)
```
